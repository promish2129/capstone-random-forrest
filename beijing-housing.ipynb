The 2nd line below is to remove the non-numeric special characters. The rest of the rows contain numeric values and removed characters left numbers only.

```python
drop_cols = ["url", "DOM", "id", "Cid", "tradeTime", "followers", "constructionTime", "renovationCondition", "district"]
drop_list = [92251,92251,92267,92270,92297,92299,92300,92304,92340,92349,92356,92235,92398,92409,92414,92467,92520,92610,92660,92814,92845,92899,113275,141376,208214,220567,220569,220570,220603,224349,243731,244054,245394]
df["floor"] = df["floor"].drop(drop_list)
df = df.dropna(subset=['floor'])
df = df.drop(drop_cols, axis=1)

df["floor"] =  [int(i) for i in df["floor"].str.replace('[^\x00-\x7F]','')]

```
