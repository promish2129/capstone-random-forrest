```python
melb_full_df = pd.read_csv("/kaggle/input/melbourne-housing-market/Melbourne_housing_FULL.csv")
melb_less_df = pd.read_csv("/kaggle/input/melbourne-housing-market/MELBOURNE_HOUSE_PRICES_LESS.csv")
melb_snap_df = pd.read_csv("/kaggle/input/melbourne-housing-snapshot/melb_data.csv")
```

Save missing columns

```python
missing_cols = []
for i in melb_full_df.columns:
    if(i not in melb_less_df):
        missing_cols.append(i)
missing_cols
```

Matching Suburb, Address, and Price from Full and Less and merge data together

```python
MatchDF = melb_full_df.merge(melb_less_df, how='left', on=list(melb_less_df.columns))
MatchDF.info()
```

```python
MatchDF1 = MatchDF.drop(["Suburb", "Address", "Method", "SellerG", "Date", "CouncilArea", "Regionname"], axis=1)
melb_snap_df = melb_snap_df.drop(["Suburb", "Address", "Method", "SellerG", "Date", "CouncilArea", "Regionname"], axis=1)
```

```python
MatchDF1.info()
```

```python
MatchDF2 = MatchDF1.merge(melb_snap_df, how='left', on=list(MatchDF1.columns))
```

```python
MatchDF2.info()
```

```python
MatchDF2.to_csv("melbourne_housing_cleaned.csv")
```
